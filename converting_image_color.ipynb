```python
from PIL import Image
import IPython.display as display

def convert_image_to_gray_bw(image_path):
    # Open the image
    img = Image.open(image_path)
    pixels = img.load()
    width, height = img.size

    # Create new images
    gray_img = Image.new("L", (width, height))
    bw_img = Image.new("1", (width, height))

    # Process each pixel
    for y in range(height):
        for x in range(width):
            r, g, b = pixels[x, y][:3]  # Get RGB values

            # Convert to grayscale
            gray_value = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
            gray_img.putpixel((x, y), gray_value)

            # Convert to black & white (threshold at 127)
            bw_value = 255 if gray_value >= 127 else 0
            bw_img.putpixel((x, y), bw_value)

    # Show images in Colab
    display.display(gray_img)
    display.display(bw_img)

# Upload an image in Colab first
from google.colab import files
uploaded = files.upload()

# Get the first uploaded file
image_path = list(uploaded.keys())[0]

# Run conversion
convert_image_to_gray_bw(image_path)

```
